### Now that the postal codes to search around have been defined, and their coordinates obtained, information on venues around them can be retrieved with the Foursquare API.

>### 5. limit on the number of search items returned (200 will be initially be used)

### Search Radius Determination

### Since we initially did know how many venues there are or how far apart the postal codes are, we did not have an initial search radius in mind.

### Arbitrary search radii could have been tested to see how many venues each one turned up, but since we still have a large number of postal codes and only a limited number of free API calls per day, we decided to use the distance between postal codes as a search radius.

### By setting the search radius to this distance, it shouldn't be so small that venues midway between postal code locations will be missed, and not so large that the search brings up too many venues or venues that are in distant neighborhoods.



## To calculate distances between postal codes, we used the geodesic function, which calculates distances between coordinates.

```python
#import geodesic function to allow distance between coordinates to be calculated

from geopy.distance import geodesic
```

## A loop was created to calculate the distance between each postal code in "df_combined_tor" and then output the results to an array.

```python
#calculate distance between the coordinates for each postal code associated with a borough that has "Toronto" in its name

dist_matrix_tor=np.zeros(shape=(df_combined_tor.shape[0],df_combined_tor.shape[0]))

for a in range(0,df_combined_tor.shape[0]):
    lat1=df_combined_tor['Latitude'][a]
    lon1=df_combined_tor['Longitude'][a]
    for b in range(0,df_combined_tor.shape[0]):
        lat2=df_combined_tor['Latitude'][b]
        lon2=df_combined_tor['Longitude'][b]
        dist_matrix_tor[a,b]=geodesic((lat1,lon1),(lat2,lon2)).m

dist_matrix_tor
```

### The array was converted to a dataframe, and the postal codes were assigned as row indices and column names for easier reading.

```python
#Convert numpy array of distances between postal code coordinates into a dataframe and format the index and column names

postal_code_tor_dist=pd.DataFrame(dist_matrix_tor)
postal_code_tor_dist.index=df_combined_tor['Postal code']
postal_code_tor_dist.columns=df_combined_tor['Postal code']
postal_code_tor_dist.reset_index(inplace=True)
postal_code_tor_dist.index.name="Index"
postal_code_tor_dist.head()
```

### The distance values in each row were then sorted from smallest to largest using the "np.sort" function.

### The lowest value in each row will be zero, which is the distance between each postal code and itself, so it is not a relevant value to consider as a search radius.

### The second lowest value in each row should be the distance from the row's respective postal code coordinates to the nearest adjacent postal code. The highest value in each row should be the distance to the most distant postal code.

### For the entire set of distances between all postal codes and their nearest neighbors, min/max/average statistics were calculated to see the range.

```python
dist_matrix_tor_sorted=np.sort(dist_matrix_tor, axis=1)
print('minimum distance between adjacent postal codes for "Toronto" boroughs only: ', dist_matrix_tor_sorted[:,1].min(), 'meters')
print('maximum distance between adjacent postal codes for "Toronto" boroughs only: ', dist_matrix_tor_sorted[:,1].max(), 'meters')
print('average distance between adjacent postal codes for "Toronto" boroughs only: ', dist_matrix_tor_sorted[:,1].mean(), 'meters')
```

#### The distances between the nearest adjacent postal codes ranged from 150 to 1993 meters, and averaged 1049 meters. The largest value would easily be within driving distance, so this distance was used for the search radius.

### The following is a function from a previous lab to extract venue category information from the Foursquare API request results.

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

### The following is the Foursquare API request code, within a for loop. The loop will run once for every postal code to explore venues around it, and the results will be output to a JSON file. The file's contents will then be normalized, and select columns of information will be appended to a new dataframe "venues_list_tor".

```python
search_radius_tor=dist_matrix_tor_sorted[:,1].max()

venues_list_tor=pd.DataFrame(columns=['Search postal code','Venue ID', 'Venue Name', 'Venue Latitude','Venue Longitude','Venue Distance from Search Point','Venue Category'])

for x in df_combined_tor.index:
    lat=df_combined_tor['Latitude'][x]
    lon=df_combined_tor['Longitude'][x]
    radius=search_radius_tor
    LIMIT=200
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)

    results_tor = requests.get(url).json()["response"]
    results_normalize_tor=json_normalize(results_tor['groups'][0]['items'])
    results_normalize_tor['venue.categories'] = results_normalize_tor.apply(get_category_type, axis=1)
    results_reduced_tor=pd.DataFrame(columns=['Search postal code','Venue ID', 'Venue Name', 'Venue Latitude','Venue Longitude','Venue Distance from Search Point','Venue Category'])
    results_reduced_tor[['Venue ID', 'Venue Name', 'Venue Latitude','Venue Longitude','Venue Distance from Search Point','Venue Category']]=results_normalize_tor[['venue.id','venue.name','venue.location.lat','venue.location.lng','venue.location.distance','venue.categories']]
    results_reduced_tor['Search postal code']=df_combined_tor['Postal code'][x]
    venues_list_tor=venues_list_tor.append(results_reduced_tor)
```

### Since the search radius applied to each postal code coordinate set overlaps with other postal code coordinates, it is likely that the venues returned in the results for each postal code coordinates may actually be closer to other postal code coordinates.

### We will calculate how far each venue is from every postal code to determine its nearest postal code. This can provide insight into the venues in the immediate area of a postal code.

### A for loop has been set up which will use the geodesic function again to calculate distances between the coordinates of each venue and every postal code. The array "venue_pc_dist_tor" output by this loop will then be input into another for loop, which will determine the closest postal code for each venue, and output that postal code's name and coordinates to another array "venue_pc_min_tor_dist."

### The information in "venue_pc_min_tor_dist" will then be inserted into the main working dataframe "working_venue_list_tor".



```python
#Determine closest postal code for each venue, which may not be the same as the postal code used for the start of the search

venue_pc_dist_tor=np.zeros(shape=(len(working_venue_list_tor.index),len(df_combined_tor.index)))

for x in working_venue_list_tor.index:
    vlat=working_venue_list_tor['Venue Latitude'][x]
    vlon=working_venue_list_tor['Venue Longitude'][x]
    for y in df_combined_tor.index:
        pclat=df_combined_tor['Latitude'][y]
        pclon=df_combined_tor['Longitude'][y]
        venue_pc_dist_tor[x,y]=geodesic((vlat,vlon),(pclat,pclon)).m

venue_pc_min_tor=[]
venue_pc_min_tor_lat=[]
venue_pc_min_tor_lon=[]
venue_pc_min_tor_dist=[]

for x in working_venue_list_tor.index:
    min_dist_index=venue_pc_dist_tor[x,:].argmin()
    venue_pc_min_tor.append(df_combined_tor["Postal code"][min_dist_index])
    venue_pc_min_tor_lat.append(df_combined_tor["Latitude"][min_dist_index])
    venue_pc_min_tor_lon.append(df_combined_tor["Longitude"][min_dist_index])
    venue_pc_min_tor_dist.append(venue_pc_dist_tor[x,:].min())

working_venue_list_tor['Venue Distance to nearest postal code']=venue_pc_min_tor_dist
working_venue_list_tor['Nearest postal code']=venue_pc_min_tor
working_venue_list_tor['Nearest postal code latitude']=venue_pc_min_tor_lat
working_venue_list_tor['Nearest postal code longitude']=venue_pc_min_tor_lon
working_venue_list_tor.head()
```

```python
#Determine the top 10 most common categories for the venues from the search results for each borough.

search_b_top_freq=[]

for i in df_results_searchb_tor_group_mean.index:
    top_count=df_results_searchb_tor_group_mean.iloc[i,1:].sort_values(ascending=False,axis=0)
    search_b_top_freq.append(list(top_count.index[0:10]))

search_b_top_freq=pd.DataFrame(search_b_top_freq)
search_b_top_freq=pd.concat([df_results_searchb_tor_group_sum[['Search Borough','Venue Count']],search_b_top_freq],axis=1)

search_b_top_freq.columns=['Search Borough','Venue Count',
                            '1st Category','2nd Category','3rd Category','4th Category','5th Category',
                            '6th Category','7th Category','8th Category','9th Category','10th Category']

print('The top 10 categories for venues in each borough search are the following:\n')
search_b_top_freq
```

### After grouping the venues by borough, the same types of categories still seem to be the most popular - coffee shops, cafes, parks, and restaurants. The only noticeable difference is in Central Toronto, where grocery stores show up as a common category.

### By venue count, it would initially appear that the Downtown Toronto has the most venues within its search area. However, remember that  many of these search results are for same venue, returned from multiple postal code searches. This count is not a reliable comparison of venue accessibility among the boroughs since higher numbers of postal codes in a borough would generally results in more venue results.

### In order to truly differentiate the number of venues within each borough's search area, duplicate venue results for each borough must be removed. The following code will remove duplicate search results (venues with the same ID, and which turned up in the searches for postal codes from the same borough).

```python
#Count the number of unique venues within the search radius of all postal codes in each borough.
borough_list=list(df_results_master_tor['Search Borough'].unique())
borough_unique_search_venues=[]

for i,y in enumerate(borough_list):
    venue_count=len(df_results_master_tor[df_results_master_tor['Search Borough']==y].drop_duplicates(subset='Venue ID'))
    borough_unique_search_venues.append(venue_count)

borough_unique_search_venues=pd.DataFrame(borough_unique_search_venues)
borough_unique_search_venues.index=borough_list
borough_unique_search_venues.reset_index(inplace=True)
borough_unique_search_venues.columns=['Search Borough','Unique Venues in Search Area']

search_borough_unique=pd.merge(search_b_top_freq[['Search Borough','Venue Count']],borough_unique_search_venues)
search_borough_unique=pd.merge(search_borough_unique,toronto_borough_pc)
search_borough_unique['Average Unique Venues/Postal Code']=search_borough_unique['Unique Venues in Search Area']/search_borough_unique['Postal code count']
search_borough_unique
```

### After the duplicate searches results were removed, it would seem that Central and Downtown Toronto actually have similar numbers of venues within their search area, even though Downtown Toronto initially returned more than twice as many venue search results as Central Toronto. Considering that Downtown Toronto has more than twice as many postal codes as Central Toronto, this disparity makes sense.

### East and West Toronto each initially had less than 1/3 of the venue search results as Downtown Toronto, but turned out well when the number of unique venues were compared.

### Downtown Toronto has the lowest average number of unique venues per postal code among the boroughs, which may indicate that there are actually relatively few unique venues in its area, or that it has a lot of postal codes that are very close together.

### A k-means clustering algorithm will be applied to the venue category proportions.

### K-means cluster size will be set to 4.

```python
#Perform k-means clustering on the venue categories, for venues within the search radius of each postal code.
#K-means cluster size set to 4

df_results_search_pc_tor_kmeans=df_results_search_pc_tor_group_mean.drop(['Search postal code'],axis=1)

kmeans1 = KMeans(n_clusters=4, random_state=0)
kmeans1.fit(df_results_search_pc_tor_kmeans)

# Add the labels to the search results to see how the venue categories in each label group compare.

kmeans1_pc_label=df_results_search_pc_tor_group_mean[['Search postal code']]
kmeans1_pc_label['K-means Cluster Label']=kmeans1.labels_
search_pc_tor_top_kmeans1=pd.merge(kmeans1_pc_label,search_pc_top_freq)
search_pc_tor_top_kmeans1.head()
```
